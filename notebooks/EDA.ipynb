```python
import math
import matplotlib.pyplot as plt
import seaborn as sns

num_cols = df.select_dtypes(include=['int64','float64']).columns

# 1. Calculate grid dimensions
num_plots = len(num_cols)
cols = 2 # Number of columns in your subplot grid
rows = math.ceil(num_plots / cols)

# 2. Create a single figure with multiple subplots
fig, axes = plt.subplots(rows, cols, figsize=(12, 4 * rows))
axes = axes.flatten() # Make it a 1D array so we can easily loop through it

# 3. Loop through columns and plot onto specific axes
for i, col in enumerate(num_cols):
    sns.boxplot(x=df[col], ax=axes[i])
    axes[i].set_title(col)

# 4. Remove any extra/empty subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout() # Ensures labels and titles don't overlap

# 5. Save the compiled figure (choose .png, .jpg, or .pdf)
fig.savefig('all_boxplots.png', dpi=300, bbox_inches='tight')
plt.show()
```
